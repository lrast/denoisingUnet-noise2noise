```python
testResults = initialTraining.forward( initialTraining.data.test_noisy )
```

```python
def imshow_custom(image):
    plt.imshow(image.permute([1,2,0]).detach())

plt.subplots(2,2)
plt.subplot(2,2,1); imshow_custom( initialTraining.data.test_noisy[0] )
plt.subplot(2,2,2); imshow_custom( testResults[0] )
plt.subplot(2,2,3); imshow_custom( initialTraining.data.test_noisy[20] )
plt.subplot(2,2,4); imshow_custom( testResults[20] )
```

The peak validation performance is achieved when the network outputs a fairly constant value. Somewhat surprising, since the UNet does have access to the ground truth at the output. Also, it is better to just pass through the noise:

```python
print('Noisy to Ground MSE:', initialTraining.pixelLoss(initialTraining.data.test_noisy, 
                                                        initialTraining.data.test_base) )
print('Reconstruction Ground MSE:', initialTraining.pixelLoss( testResults,
                                                        initialTraining.data.test_base) )
```

This looks like about the right reconstruction difference considering the validation performance. It does beat not trying to denoise at all, but not by much.

It is a bit surprising that this is the best that you can do. In particular, the colors of the reconstructions are all very dull. I wonder if this is due to the [0,1] encoding that I''m forcing on the network. Maybe a nonlinearity would be better.

## Noise free reproduction

```python
import pandas as pd

baseLine = models.UNet(noise_rate=0.0)
baseLine.setup()

metrics = {
            'MSE':torch.nn.MSELoss(),
            'BCE':torch.nn.BCELoss(),
            'L1':torch.nn.L1Loss()
        }
   
    
outs = []

for i in range(3):
    for model in ['BCE', 'MSE', 'L1']:
        currModel = models.UNet.load_from_checkpoint( 'lightning_logs/noiseFree/'+ model + str(i) + '/best.ckpt')
        
        for metricName, L in metrics.items():
            outs.append( 
                [model, metricName, L(currModel.forward(baseLine.data.test_base), baseLine.data.test_base).detach().numpy()] 
            )
        
        
results = pd.DataFrame(outs, columns=['model', 'metric', 'loss'])
```

```python
figsize(30,10)
f, axes = plt.subplots(1, 3)
plt.rcParams.update({'font.size':32})
sns.stripplot(data=results[ results.metric=='BCE'] , x='model', y='loss', ax=axes[0], size=14); axes[0].set_title('BCE Loss')
sns.stripplot(data=results[ results.metric=='MSE'] , x='model', y='loss', ax=axes[1], size=14); axes[1].set_title('MSE Loss')
sns.stripplot(data=results[ results.metric=='L1'] , x='model', y='loss', ax=axes[2], size=14); axes[2].set_title('L1 Loss')
```

```python
def imshow_custom(image):
    plt.imshow(image.permute([1,2,0]).detach())

figsize(8, 30)
plt.subplots(10,4)

BCEresults = models.UNet.load_from_checkpoint( 'lightning_logs/noiseFree/BCE2/best.ckpt').forward( baseLine.data.test_base)
MSEresults = models.UNet.load_from_checkpoint( 'lightning_logs/noiseFree/MSE2/best.ckpt').forward( baseLine.data.test_base)
L1results = models.UNet.load_from_checkpoint( 'lightning_logs/noiseFree/L12/best.ckpt').forward( baseLine.data.test_base)


for i in range(1, 11):
    base = 4*(i-1) + 1
    plt.subplot(10,4,base)
    ax = plt.gca(); ax.set_xticks([]); ax.set_yticks([])
    ax.set_title(i)
    imshow_custom( baseLine.data.test_base[i] )
    
    plt.subplot(10,4,base+1)
    ax = plt.gca(); ax.set_xticks([]); ax.set_yticks([])
    ax.set_title('BCE')
    imshow_custom( BCEresults[i] )
    
    plt.subplot(10,4,base+2)
    ax = plt.gca(); ax.set_xticks([]); ax.set_yticks([])
    ax.set_title('MSE')
    imshow_custom( MSEresults[i] )
    
    plt.subplot(10,4,base+3)
    ax = plt.gca(); ax.set_xticks([]); ax.set_yticks([])
    ax.set_title('L1')
    imshow_custom( L1results[i] )
    
```

### Visually, they all look OK.

I generally like the L1 results the best, in particular for the example shown here ^, but also across different training samples. 

- L1 also seems to be generally the most consistent across training runs, which is a good thing.


In many cases, regardless of objective function, the colors seem off, which is quite surprising considering the lack of noise.

- However, The L1 objective function models also trained for the longest (early stopping), compared to the other models, and in several cases the training appears to see continued improvement, even when the early stopping doesn't catch it. The other objective functions may simply be 'noisier' during training

#### This leads me to favor:
1. higher patience for future training
2. L1 loss
